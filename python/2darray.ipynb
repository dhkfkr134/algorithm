### 2차원 배열

#### 행렬 덧셈 - 2738

```python
x,y = map(int,input().split())
arr = [list(map(int,input().split())) for _ in range(x)]
arr2 = [list(map(int,input().split())) for _ in range(x)]

for i in range(x):
    for j in range(y):
        print(arr[i][j] + arr2[i][j], end = " ")
    print()
```

#### 세로읽기 - 10798

```python
행, 열 바꾸기
arr = []
maxlen = 0
for i in range(5):
    arr.append(list(input()))
    maxlen = max(len(arr[i]),maxlen)
for y in range(maxlen):
    for x in range(5):
        if len(arr[x]) > y:
            print(arr[x][y], end="")
```
