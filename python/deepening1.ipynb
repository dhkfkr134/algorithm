#### 그룹 단어 체커 - 1316

```python
# 단어마다 포인터 두개로 점프했는지 체크하고 점프했는데 다시 내려오면 그룹단어에서 빼는 방식
# aaaaa (점프).b.c.d... a(착지))
def solution(n, group):
    answer = n
    for word in group:
        flag = False
        for i in range(len(word)):
            isjumped = False
            if flag:
                break
            for j in range(i, len(word)):
                if isjumped:
                    if word[i] == word[j]:
                        answer = answer - 1
                        flag = True
                        break
                else:
                    isjumped = True if word[i] is not word[j] else False
    return answer

n = int(input())
group = [ input() for _ in range(n)]
print(solution(n,group))
```

```python
# 파이썬다운 풀이
# 검사하려는 문자가 뒷 문자와 같은지 확인하고, 그 뒤에 이어지는 문자열에도 있는지 확인한다.
N = int(input())
cnt = N

for i in range(N):
    word = input()
    for j in range(0, len(word)-1):
        if word[j] == word[j+1]:
            pass
        elif word[j] in word[j+1:]:
            cnt -= 1
            break
print(cnt)
```
