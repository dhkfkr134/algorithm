#### baekjoon5622

```python
string = input()
sum = 0
for s in string:
    x = (ord(s) - ord('A') + 9)
    if x < 24:
        sum += int(x/3)
    elif 24 <= x & x < 28:
        sum += 8
    elif 28 <= x & x < 31:
        sum += 9
    elif 31 <= x & x < 35:
        sum += 10
print(sum)
```

입력값 그대로 출력하기.  
코딩테스트는 입출력값을 file형태로 갖고있기때문에 EOF가 존재한다.
콘솔로 입력값을 넣으면 무한으로 \n을 넣을 수 있음을 유의하자.

```python
while True :
    try :
        print(input())
    except EOFError:
        break
```
