## 문제 58. 리스트 정렬
> 연결 리스트를 O(n logn)에 정렬하라.

* 입력 : head = [ 4 2 1 3 ]
* 출력 : [ 1 2 3 4 ]

```python
# 58. 리스트 정렬

from typing import List

# 링크드 리스트 데이터타입 정의
class ListNode:
    def __init__(self, x=0, next=None):
        self.val = x
        self.next = None

    def print(self):
        curn = self
        string = ""
        while curn:
            string += str(curn.val)
            if curn.next:
                string += "->"
            curn = curn.next
        print(string)


class Solution:
    # 두 정렬 리스트 병합
    def merge2Lists(self, l1: ListNode, l2: ListNode) -> ListNode:
        if l1 and l2:
            if l1.val > l2.val:
                l1, l2 = l2, l1
            l1.next = self.merge2Lists(l1.next, l2)

        return l1 or l2

    def sortList(self, head: ListNode) -> ListNode:
        if not (head and head.next):
            return head

        # 런너 기법 활용
        half, slow, fast = None, head, head
        while fast and fast.next:
            half, slow, fast = slow, slow.next, fast.next.next
        half.next = None

        # 분할 재귀 호출
        l1 = self.sortList(head)
        l2 = self.sortList(slow)

        return self.merge2Lists(l1, l2)
    # 리스트에서 링크드리스트로
    def list2node(self, list1: List) -> ListNode:
        result_node = ListNode()
        
        for i,num in enumerate(list1):
            if i == 0 :
                result_node.val = num
            else :
                node = result_node
                while node.next != None:
                    node = node.next
                node.next = ListNode(num)
        return result_node
```

```python
S = Solution()
head = S.list2node([1, 5, 3, 4, 0])
S.sortList(head).print()
```

```python
from typing import List


class Solution:
    def merge(self, intervals: List[List[int]]) -> List[List[int]]:
        merged = []
        for i in sorted(intervals, key=lambda x: x[0]):
            print(intervals)
            if merged and i[0] <= merged[-1][1]:
                merged[-1][1] = max(merged[-1][1], i[1])
            else:
                merged += i,
        return merged
```

## 문제 60. 삽입 정렬 리스트
> 연결 리스트를 삽입 정렬로 정렬하라.

* 입력 : head = [ 4 2 1 3 ]
* 출력 : [ 1 2 3 4 ]

```python
class ListNode:
    def __init__(self, x=0):
        self.val = x
        self.next = None

    def print(self):
        
        curn = self
        string = ""
        while curn:
            string += str(curn.val)
            if curn.next:
                string += "->"
            curn = curn.next
        print(string)    

class Solution:
    def list2node(self, list1: List) -> ListNode:
        result_node = ListNode()
            
        for i,num in enumerate(list1):
            if i == 0 :
                result_node.val = num
            else :
                node = result_node
                while node.next != None:
                    node = node.next
                node.next = ListNode(num)
        return result_node

    def insertionSortList(self, head: ListNode) -> ListNode:
        cur = parent = ListNode(None)
        i = 0
        while head:
            print("@@while문실행중")
            while cur.next and cur.next.val < head.val:
                print("@@@@두번째실행중")
                cur = cur.next

            cur.next, head.next, head = head, cur.next, head.next
            i+=1
            print(i,"//")
            cur.print()
            if head is not None:
                head.print()

            cur = parent
        return cur.next

    def fixedInsertionSortList(self, head: ListNode) -> ListNode:

        cur = parent = ListNode(0)
        while head:
            while cur.next and cur.next.val < head.val:
                cur = cur.next

            cur.next, head.next, head = head, cur.next, head.next

            # 필요할때만 cur = parent처리
            if head and cur.val > head.val:
                cur = parent
        return parent.next
```

```python
S = Solution()
head = S.list2node([4,2,1,3])
S.insertionSortList(head).print()
```
